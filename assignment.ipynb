```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

Random data generated w/ Pandas and Numpy

```python
np.random.seed(42)
data = pd.DataFrame({
    'A': np.random.randn(50),
    'B': np.random.randn(50),
    'C': np.random.randn(50),
    'D': np.random.randn(50)
})
```

My scatter plot generated w/ Seaborn and Matplotlib

```python
sns.pairplot(data)
plt.suptitle('Pair Plot of Variables', y=1.02)
plt.show()
```

My heat map generated w/ Seaborn and Matplotlib

```python
corr = data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap of Variables')
plt.show()
```
